```python
# Analyse loss plots

import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import os 
import sys

# dir_path = os.path.dirname(os.getcwd())
# data_path = os.path.join(dir_path, 'data', 'trained_models', 'new_models')
# sys.path.append(data_path)

data = '/home/sirfuser/shared/no_relu/median/model_m_UNet_n10000_e1000_lr1e-05_b32_s0.8_0.1_r0_gFalse/losses.csv'
```

```python
losses = pd.read_csv(data)
plt.figure(figsize=(10, 6))
plt.plot(losses['train_loss'], label='Training Loss', color='blue', linewidth=2)
plt.plot(losses['valid_loss'], label='Validation Loss', color='orange', linewidth=2)

# Labels and title
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim(0, 100)
plt.ylim(0, 0.01)
plt.title('Training vs Validation Loss for First 100 Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout
plt.savefig(f"/home/sirfuser/devel/PET_sensitivity/data/training_plots/median_loss_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import os 
import sys

dir_path = os.path.dirname(os.getcwd())
data_path = os.path.join(dir_path, 'data', 'trained_models', 'new_models')
sys.path.append(data_path)

data = os.path.join(data_path,'model_m_UNet_n10000_e1000_lr0.0001_b32_s0.8_r0_gFalse/losses.csv')


losses = pd.read_csv(data)
plt.figure(figsize=(10, 6))
plt.plot(losses['train_loss'], label='Training Loss', color='blue', linewidth=2)
plt.plot(losses['valid_loss'], label='Validation Loss', color='orange', linewidth=2)

# Labels and title
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.title('Training vs Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
